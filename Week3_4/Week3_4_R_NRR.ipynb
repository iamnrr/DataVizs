```python
# Import required libraries
import pandas as pd
# Treemap Ploting
import squarify
# Matplotlib and Seaborn imports
import matplotlib
from matplotlib import style
import matplotlib.pyplot as plt
import seaborn as sns
# Activate Seaborn
sns.set()
%matplotlib inline

import pandas as pd
```

# Tree Maps

```python
# Reading CSV file
Treemapdata = pd.read_csv("Treemapdata.csv")
Treemapdata.head()
```

```python
# make plot
fig = plt.figure(figsize=(12, 16))
fig.suptitle("Revenues for Topics", fontsize=20)

cmap=matplotlib.cm.viridis

norm = matplotlib.colors.Normalize(vmin=min(Treemapdata.REVENUE), vmax=max(Treemapdata.REVENUE))
colors = [cmap(norm(value)) for value in Treemapdata.REVENUE]

#ax = fig.add_subplot(111, aspect="equal")
ax = squarify.plot(sizes=Treemapdata['REVENUE'], label=Treemapdata['TOPIC'], color = colors, alpha=.8 )
ax.set_xticks([])
ax.set_yticks([])

colors = plt.imshow([Treemapdata.REVENUE], cmap=matplotlib.cm.viridis) #matplotlib.cm.Blues(norm(value)))
fig.colorbar(colors, orientation="vertical", shrink=.96)

plt.show()
```

# Area Charts

```python
areadata = pd.read_excel("us-postage.xlsx")
areadata.head()
```

```python
fig = plt.figure(figsize=(12, 16))

# Area plot
plt.fill_between(areadata.Year, areadata.Price)
plt.title("US Postage Prices over Years")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

# Stacked Area Charts

```python
stareadata = pd.read_excel("obama-approval-ratings.xlsx")
stareadata.head()
```

```python
# Make the plot
plt.figure(figsize=(12, 16))
plt.stackplot(stareadata.Issue,  stareadata["Approve"],  stareadata["Disapprove"],  stareadata["None"], 
              labels=['Approve','Disapprove','None'])
plt.legend(loc='upper left')
plt.margins(0,0)
plt.title('Obama approval ratings')
plt.xlabel("Issues")
plt.ylabel("Ratings")
plt.xticks(rotation=90)
plt.show()
```

# Step Charts

```python
stepdata = pd.read_excel("us-postage.xlsx")
stepdata.head()
```

```python
# plot out the options
fig = plt.figure(figsize=(12, 16))
plt.plot(stepdata.Year, stepdata.Price, drawstyle='steps', linestyle='-', label='steps-pre aka steps', alpha=0.5,)
plt.title("US Postage Prices over Years")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```
