```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
# Seaborn for plotting and styling
import seaborn as sns
```

# Line Chart

```python
worldpop_data =  pd.read_excel('world-population.xlsx', sheetname='world-population')
#worldpop_data.head()
```

```python
fig, ax = plt.subplots(figsize = (20,6))
plt.figure()

ax.plot(worldpop_data['Year'], worldpop_data['Population'], color = 'green')
ax.set_title("World Population - Trend")
ax.set_xlabel('Year')
ax.set_ylabel('Population')

ax.set_xticks(worldpop_data['Year'])
ax.set_xticklabels(worldpop_data['Year'], rotation=45)

fig = plt.gcf()
ax.figure.savefig('LineChart_python.png')

plt.show()
```

# Bar Chart

```python
obama_data =  pd.read_excel('obama-approval-ratings.xlsx', sheetname='Sheet1')
obama_data
```

```python
fig, ax = plt.subplots(figsize = (12,6))
plt.figure("python_barchart.png")
python_barchart = sns.barplot(x=obama_data['Issue'], y = obama_data['Approve'],  
                        data = obama_data,
                        ax=ax, ci=None)

ax.set_title("Obama Approval ratings for all Issues")
ax.set_ylim(0,100)
ax.set_xticklabels(list(obama_data['Issue']), rotation=45)

fig = plt.gcf()
ax.figure.savefig('BarChart_python.png')
```

# Stacked Bar Chart

```python
obama_data.head()
```

```python
Issue = obama_data['Issue']
approve = obama_data['Approve']
disapprove = obama_data['Disapprove']
none = obama_data['None']
index = np.arange(len(obama_data))
```

```python
np.asarray(list(obama_data['Issue']))
```

```python
plt.figure(figsize = (16,12), dpi=100)
approvebar = plt.bar(x= index, height = obama_data['Approve'], width=1)
disapprovebar = plt.bar(x= index, height = obama_data['Disapprove'], width=1, bottom =  obama_data['Approve'])
nonebar = plt.bar(x= index, height = obama_data['None'], width=1, 
                  bottom =  obama_data['Approve']+obama_data['Disapprove'])

plt.ylabel('Ratings')
plt.title('Obama approval ratings - Issues')
plt.xticks(index, Issue, rotation=45)
plt.legend((approvebar[0], disapprovebar[0], nonebar[0]), ('Approve', 'Disapprove','None'), # leged mapping
           bbox_to_anchor=(1, 1),loc=2)  # to save legend on the right
 
# saving the chart
plt.savefig('stacked_barchart_python.png')

# Show graphic
plt.show()
```

# Pie Chart

```python
obama_data2 =  pd.read_excel('obama-approval-ratings-pie.xlsx', sheetname='Sheet1')
obama_data2.head()
```

```python
obama_data_pie = obama_data2[obama_data2['Issue'] == 'Foreign Affairs']
obama_data_pie
```

```python
plt.figure(figsize = (12,8), dpi=100)

plt.pie(obama_data_pie['Percentage'], labels = obama_data_pie['Status'], autopct = '%.f')

#plt.ylabel('Ratings')
plt.title('Obama approval ratings - for Foreign Affairs')
#plt.xticks(index, Issue, rotation=45)
plt.legend()
 
# saving the chart
plt.savefig('piechart_python.png')

# Show graphic
plt.show()
```

# Donut Charts

```python
plt.figure(figsize = (12,8), dpi=100)


# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.7, color='white')


# Give color names
plt.pie(obama_data_pie['Percentage'], labels = obama_data_pie['Status'], 
        colors=['orange','lightgreen','skyblue'], autopct = '%.f')
p=plt.gcf()
plt.title('Obama approval ratings - for Foreign Affairs')
plt.legend()
p.gca().add_artist(my_circle)

# saving the chart
plt.savefig('donutchart_python.png')

# Show graphic
plt.show()
```
